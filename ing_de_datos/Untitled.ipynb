```python
import bs4
import requests

response = requests.get('https://www.elnacional.com/')

soup = bs4.BeautifulSoup(response.text, "html.parser")

links = soup.select('.item-details a')

titulos = [link.text for link in links]

i=1
for titular in titulos:
    print('{}. {}'.format(i,titular))
    i+=1
    
```

```python
response3 = requests.get("https://www.theguardian.com/international")

soup3 = bs4.BeautifulSoup(response3.text,'html.parser')

noticias_titulos = soup3.select('.fc-item__container .fc-item__content .fc-item__header .fc-item__title .fc-item__link .u-faux-block-link__cta .js-headline-text')

#print(dir(noticias_titulos[1]))
#print(noticias_titulos[1].text)

noticias_titulos_text = [noticia.text for noticia in noticias_titulos]
i=0
for noticia in noticias_titulos_text:
    print(str(i)+": " + noticia+"\n\n")
    i+=1
```
